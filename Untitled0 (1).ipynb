# Imports

```python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, InputLayer, Dropout
from tensorflow.keras.layers.experimental import preprocessing
from tensorflow.keras.applications import VGG16
from tensorflow.keras.applications import MobileNetV3Small
from tensorflow.keras.callbacks import EarlyStopping
import random
import numpy as np
from sklearn.metrics import accuracy_score
```

# Dataset und Validierungsset erstellen

```python
# Verzeichnis mit den Bildern
data_dir_train = '/content/drive/MyDrive/eigenesSet/train_four_label'

# Datengeneratoren erstellen
datagen = ImageDataGenerator(rescale=1/255.,
    #rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest')

# Training Datengenerator
train_generator = datagen.flow_from_directory(
    data_dir_train,
    target_size=(240, 320),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

class_indices = train_generator.class_indices

# Print the class indices
print("Class indices:", class_indices)
```

```python
# Get the unique class labels from the generator
selected_class = 'blue'  # Replace with the class name you want to sample

# Get the indices of images belonging to the selected class
class_indices = np.where(train_generator.classes == train_generator.class_indices[selected_class])[0]

# Randomly sample 20 indices from the selected class
sampled_indices = random.sample(list(class_indices), 20)

# Set up a figure to display sampled images
fig, axs = plt.subplots(2, 10, figsize=(15, 5))

# Iterate through each sampled image index
for i, image_index in enumerate(sampled_indices):
    # Get the image and label from the current batch
    image = train_generator[0][0][image_index % train_generator.batch_size]
    label = train_generator[0][1][image_index % train_generator.batch_size]

    # Convert one-hot encoded label to class name
    class_name = list(train_generator.class_indices.keys())[int(label.argmax())]

    # Display the image
    axs[i // 10, i % 10].imshow(image)
    axs[i // 10, i % 10].set_title(f"Class: {class_name}")
    axs[i // 10, i % 10].axis('off')

plt.show()
```

# Validierungsset

```python
# Verzeichnis mit den Bildern
data_dir_eval = '/content/drive/MyDrive/eigenesSet/test_four_label'

# Datengenerator für Evaluierung erstellen
eval_datagen = ImageDataGenerator(rescale=1/255.)

# Evaluierung Datengenerator
eval_generator = eval_datagen.flow_from_directory(
    data_dir_eval,
    target_size=(240, 320),  # Achte darauf, dass die Größe dem Modell entspricht
    batch_size=32,
    class_mode='categorical',
    shuffle=False  # Setze shuffle auf False, damit die Reihenfolge beibehalten wird
)
class_indices = eval_generator.class_indices

# Print the class indices
print("Class indices:", class_indices)
```

```python
def plot_images(images, labels, class_labels):
    plt.figure(figsize=(15, 8))
    for i in range(len(images)):
        plt.subplot(4, 8, i + 1)
        plt.imshow(images[i])
        plt.title(class_labels[np.argmax(labels[i])])
        plt.axis("off")
    plt.show()

# Get a batch of images and labels from the evaluation generator
images, labels = next(eval_generator)

# Convert labels to class labels
class_labels = list(eval_generator.class_indices.keys())

# Plot the images with labels
plot_images(images, labels, class_labels)
```

# CNN Model erstellen

```python
model = Sequential()
model.add(Conv2D(16, (3, 3), activation='relu', input_shape=(240, 320, 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(32, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
#model.add(Dropout(0.1))
model.add(Dense(4, activation='softmax'))

#early stop
early_stopping = EarlyStopping(monitor='val_accuracy', patience=10, restore_best_weights=True)
# Modell kompilieren
opt = keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
```

# pre trained model


```python
# Load pre-trained VGG16 model
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(240, 320, 3))

# Freeze the layers of the pre-trained model
for layer in base_model.layers:
    layer.trainable = False

# Create a new model on top of the pre-trained model
model2 = Sequential()
model2.add(base_model)
model2.add(Flatten())
model2.add(Dense(16, activation='relu'))
model2.add(Dropout(0.1))
model2.add(Dense(4, activation='softmax'))  # Adjust the number of output classes

early_stopping = EarlyStopping(monitor='val_accuracy', patience=10, restore_best_weights=True)

# Compile the model
model2.compile(optimizer=keras.optimizers.Adam(learning_rate=0.0001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

```

# Model trainieren

```python
model.fit(train_generator, epochs=200, validation_data=eval_generator, callbacks=[early_stopping])
```

# evaluieren

```python
def plot_accuracy_with_incrementing_validation(model, eval_generator, increment_step=0.05, max_size=1.0):
    accuracies = []
    val_sizes = []

    total_samples = eval_generator.samples
    max_val_size = int(total_samples * max_size)
    print(total_samples)
    current_size = int(0.05 * total_samples)  # Start with 5% of the evaluation set
    print(current_size)
    while current_size <= max_val_size:
        eval_generator.reset()
        eval_generator.index_array = eval_generator.index_array[:current_size]

        # Use model.evaluate to get accuracy
        loss, accuracy = model.evaluate(eval_generator)
        accuracies.append(accuracy)
        val_sizes.append(current_size / total_samples)
        print(current_size)

        current_size = int(current_size + total_samples * increment_step)

    plt.plot(val_sizes, accuracies, marker='o')
    plt.title('Accuracy vs. Validation Set Size')
    plt.xlabel('Validation Set Size (as a percentage)')
    plt.ylabel('Accuracy')
    plt.show()

# Example usage:
# Assuming you have a model named 'your_model'
plot_accuracy_with_incrementing_validation(model2, eval_generator)
```

```python
def plot_losses(history):
    # Extract training and validation loss from the history object
    training_loss = history.history['accuracy']
    validation_loss = history.history['val_accuracy']

    # Plotting the training and validation loss
    epochs = range(1, len(training_loss) + 1)
    plt.plot(epochs, training_loss, 'bo-', label='Training acc')
    plt.plot(epochs, validation_loss, 'ro-', label='Validation acc')

    # Customize the plot
    plt.title('Training and Validation acc')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    # Show the plot
    plt.show()

plot_losses(model2.history)
```
